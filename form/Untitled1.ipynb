```python
from pronsole import pronsole
printer=pronsole()
printer.do_load(path)
print(printer.all_layers_for_opengl)
```

```python

for i in range(len(printer.fgcode.all_layers)-1):
    print(printer.fgcode.all_layers[i][0].current_z)
print(printer.fgcode.duration)
```

```python
all_layer_lines=[]
for i in range(2,len(printer.fgcode.all_layers)-2):
    lines=[]
    tmp_line=None
    line_len=len(printer.fgcode.all_layers[i])
    for j in range(line_len):
        if printer.fgcode.all_layers[i][j].command=='G0' and printer.fgcode.all_layers[i][j].x and printer.fgcode.all_layers[i][j].y:
            if tmp_line:
                lines.append(tmp_line)
            tmp_line=[]
            tmp_line.append((printer.fgcode.all_layers[i][j].x,printer.fgcode.all_layers[i][j].y))
        if printer.fgcode.all_layers[i][j].command=='G1' and  printer.fgcode.all_layers[i][j].x and printer.fgcode.all_layers[i][j].y:
            tmp_line.append((printer.fgcode.all_layers[i][j].x,printer.fgcode.all_layers[i][j].y))
        if j==line_len-1:
            lines.append(tmp_line)
    all_layer_lines.append(lines)
print(all_layer_lines)
print(len(all_layer_lines))
```
